# 02. Data Wrangling

### Renaming Columns

```python
# Renaming columns 'Surnam', 'n_dependants' and 'fam_status'
df.rename(columns = {'Surnam':'Surname', 'n_dependants':'number_of_dependants', 'fam_status':'family_status'}, inplace = True)
```

```python
# Checking to see output
df.head()
```

### Changing Data Types

```python
# Changing data type of 'user_id', 'Age', 'number_of_dependants' and 'income'
df['user_id'] = df['user_id'].astype('str')
df['Age'] = df['Age'].astype('int8')
df['number_of_dependants'] = df['number_of_dependants'].astype('int8')
df['income'] = df['income'].astype('int32')
```

```python
# Checking output
df.info()
```

### Dropping Columns

```python
# Dropping columns 'First Name' and 'Surname', because they are personally identifiable information (PII)
df = df.drop(columns = ['First Name', 'Surname'])
```

```python
# Checking output
df.head()
```

# 03. Data Consistency Checks

### Mixed Data Type

```python
# Checking for mixed data type
for col in df.columns.tolist():
  weird = (df[[col]].applymap(type) != df[[col]].iloc[0].apply(type)).any(axis = 1)
  if len (df[weird]) > 0:
    print (col)
```

```python
# Addressing the column with mixed data type
df['First Name'] = df['First Name'].astype('str')
```

```python
# Checking output of mixed data type fix
for col in df.columns.tolist():
  weird = (df[[col]].applymap(type) != df[[col]].iloc[0].apply(type)).any(axis = 1)
  if len (df[weird]) > 0:
    print (col)
```

### Importing prepared Instacart data

```python
# Import orders_products_updated_4_8a.pkl as ords_prods_merged
ords_prods_merged = pd.read_pickle(os.path.join(path, '02 Data', 'Prepared Data', 'orders_products_updated_4_8a.pkl'))
```

#### There are some columns in the merged dataframe that are not necessary for the analysis. These will be dropped.

```python
# Dropping unnecessary columns
ords_prods_merged = ords_prods_merged.drop(columns = ['new_user_order', '_merge 1', '_merge'])
```

#### The data type of 'user_id' will also be changed to 'str'

```python
# Changing data type 'user_id' of ords_prods_merged to 'str'
ords_prods_merged['user_id'] = ords_prods_merged['user_id'].astype('str')
```

### Merging Dataframes

```python
# Merging df with ords_prods_merged on 'user_id'
df_merged = df.merge(ords_prods_merged, on = 'user_id', indicator = True)
```
